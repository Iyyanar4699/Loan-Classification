```python
Df_cleaned = Df.dropna(subset=['LoanAmount'])
```

```python
Mean_Loan_Term=Df_cleaned['Loan_Amount_Term'].mean()
```

```python
Df_cleaned['Loan_Amount_Term'].fillna(Mean_Loan_Term, inplace=True)
```

```python
Df_cleaned['Credit_History'].fillna(1.0, inplace=True)
```

```python
Df_cleaned['Gender'].fillna('Male', inplace=True)
```

```python
Df_cleaned['Dependents'] = Df_cleaned['Dependents'].replace('3+', '3')
```

```python
Df_cleaned.isnull().sum()
```

```python
Df_cleaned['Dependents'].fillna('0', inplace=True)
```

```python
Df_cleaned['Self_Employed'].fillna('No', inplace=True)
```

```python
Df_cleaned['Married'].fillna('Yes', inplace=True)
```

```python
Df_cleaned.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
Label=LabelEncoder()
Df_cleaned['Gender'] = Label.fit_transform(Df_cleaned['Gender'])
Df_cleaned['Education'] = Label.fit_transform(Df_cleaned['Education'])
Df_cleaned['Property_Area'] = Label.fit_transform(Df_cleaned['Property_Area'])
```

```python
Df_Onehot=pd.get_dummies(Df_cleaned, columns=['Married','Self_Employed'])
columns_to_convert = ['Married_No', 'Married_Yes', 'Self_Employed_No', 'Self_Employed_Yes']
Df_Onehot[columns_to_convert] = Df_Onehot[columns_to_convert].astype(int)
```

```python
Df_Onehot
```

```python
x=Df_Onehot.drop(['Loan_Status','Loan_ID'], axis=1)
y=Df_Onehot['Loan_Status']
```

```python
y=Label.fit_transform(Df_Onehot['Loan_Status'])
y = y.reshape(-1, 1)
y=np.ravel(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import tensorflow as tf
```

```python
ann=tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=13,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=12,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(X_train,y_train,batch_size=32,epochs=50)
```

```python
y_pred=ann.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
y_test = y_test.astype(int)
y_pred = y_pred.astype(int)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear', random_state=42)
model.fit(X_train, y_train)
```

```python
y_Pred = model.predict(X_test)
```
